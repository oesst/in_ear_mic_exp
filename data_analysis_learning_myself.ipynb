## This notebook shows the result for the learning experiment with myself.
Only rippled noise was presented in monaural as well as binaural condition. Both condition had 300 trials. Experiment started with binaural condition

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


class LinearReg():

    def __init__(self, x, y):
        from sklearn.linear_model import LinearRegression

        self.lr_model = LinearRegression()

        self.x = x.reshape(-1, 1)
        self.y = y.reshape(-1, 1)

        self.lr_model.fit(self.x, self.y)

        self.rr = self.lr_model.score(self.x, self.y)

    def get_fitted_line(self):
        return [self.x, self.lr_model.predict(self.x)]

    def get_coefficients(self):
        return self.lr_model.coef_[0, 0], self.lr_model.intercept_[0]

    def get_score(self, x=0, y=0):
        if x == 0 or y == 0:
            return self.rr
        else:
            return self.lr_model.score(x, y)

    def print_coefficients(self):
        print('Gain: {0:1.2f}, Bias: {1:1.2f}, , r^2: {2:1.2f}'.format(self.lr_model.coef_[0, 0], self.lr_model.intercept_[0], self.rr))
        return ('Gain: {0:1.2f},\nBias: {1:1.2f},\n' + r'$r^2$: {2:1.2f}').format(self.lr_model.coef_[0, 0], self.lr_model.intercept_[0], self.rr)



```

## Learning (Error over trials)

Add new column `error` with `abs(line_number - user_estimate`). Additionally, merge columns `sound_type` and `condition` in column `condition`. Additionally, rename conditions and create a new DataFrame with all the data merged

```python
df.head()
```

```python
# dfs_merged = pd.concat(dfs)


for i_df,df in enumerate(dfs):
    
    # merged conditions and sounds
    df['condition_merged'] = df[df.columns[4:5]].apply(lambda x : '-'.join(x.astype(str)),axis=1)
    # calculate abs error
    df['error'] = np.abs(df.user_estimate-df.line_number)
    # rename the new conditions
    df.replace({'condition_merged' : { 'rippled-bin' : 'bin rippled', 'rippled-mono' : 'mono rippled', 'white-bin' : 'bin white', 'white-mono' : 'mono white'}},inplace=True)
    # replace df
    dfs[i_df] = df
    
# new data frame
dfs_merged = pd.concat(dfs)
```

```python
from scipy.optimize import curve_fit

def func(x, a, b, c):
    return a * np.exp(-x/ b) + c



fig = plt.figure(figsize=(20,5))
axes = fig.subplots(len(dfs),2,squeeze=False,sharex=True,sharey=True)

                
for i_df,df in enumerate(dfs):
        
    for i_con,con in enumerate(np.sort(pd.unique(df.condition_merged))[[1,0]]):
        
        ax = axes[i_df,i_con]
        
        if i_con == 0 :
            ax.set_ylabel('Par #{}\n'.format(i_df))
        
        if i_con == 0 and i_df == int(len(dfs)/2):
            ax.set_ylabel('Par #{} \nAbs(error)'.format(i_df))

        
        if i_df == 0:
            ax.set_title(con)
            
        if i_df == len(dfs)-1:
            ax.set_xlabel('Trials')
        
        
        err = df.error[df.condition_merged == con]
        
        ax.plot(np.arange(0,len(err)),err)
        
        # plot fitted regression line
        x = np.arange(0,len(err))

        popt, pcov = curve_fit(func, x, err)

        ax.plot(x, func(x, *popt), '-', color='C3', label="Fitted Curve")

            


plt.tight_layout()


# plt.savefig('results_learning.svg', dpi=300, bbox_inches='tight')
# plt.savefig('results_learning.png', dpi=300, bbox_inches='tight')

plt.show()
```
